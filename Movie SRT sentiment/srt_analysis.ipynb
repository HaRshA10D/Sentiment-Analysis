```python
from textblob import TextBlob
```

```python
x = 0.0
i = 0
for sub in subs:
    subt = TextBlob(sub.text)
    if subt.sentiment.polarity != 0:
        print(subt.sentiment.polarity)
        x = x + subt.sentiment.polarity
        i++

total = x/float(i)
print(total)
```
